```python
data1 = [i for i in range(0,10)]
data2 = [i**2 for i in range(0,10)]
data3 = [i**3 for i in range(0,10)]
plt.plot(data1,data1, label = "linear")
plt.plot(data1,data2, label = "square")
plt.plot(data1,data3, label = "cubic")
plt.legend()
plt.title("Comparison of plots")
plt.show()
```

```python
data1 = [i for i in range(0,10)]
data2 = [i**2 for i in range(0,10)]
data3 = [i**3 for i in range(0,10)]
plt.plot(data1,data1, label = "linear")
plt.plot(data1,data2, label = "square")
plt.plot(data1,data3, label = "cubic")
plt.legend()
plt.title("Comparison of plots")
plt.show()
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize = (10,6))

ax1.plot(data1,data1)
ax1.set_title("linear")
ax1.set_xlabel("x_values")
ax1.set_ylabel("y_values")

ax2.plot(data1,data2)
ax2.set_title("quadratic")
ax2.set_xlabel("x_values")
ax2.set_ylabel("y_values")

plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,10))

axs[0,0].plot(data1,data1)
axs[0,0].set_title("linear")
axs[0,0].set_xlabel("x_values")
axs[0,0].set_ylabel("y_values")

axs[0,1].plot(data1,data2)
axs[0,1].set_title("quadratic")
axs[0,1].set_xlabel("x_values")
axs[0,1].set_ylabel("y_values")

axs[1,0].plot(data1,data3)
axs[1,0].set_title("cubic")
axs[1,0].set_xlabel("x_values")
axs[1,0].set_ylabel("y_values")

axs[1,1].bar(x,y)



plt.tight_layout()
plt.show()
```
